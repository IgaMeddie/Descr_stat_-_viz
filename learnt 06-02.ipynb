# Simple graph

```python
x = np.linspace(0, 1, 60) # X axis - 60 evenly spaced numbers between 0 and 1

y = np.random.randn(len(x)).cumsum()


'''np.random.randn(len(x)): This generates an array of random numbers drawn from a standard normal distribution (mean=0, standard deviation=1), with the length of the array being the same as the length of the array x.

.cumsum(): This method is then applied to the array of random numbers, calculating the cumulative sum of the elements. In other words, it computes the running total of the elements in the array.'''
```

```python
plt.plot(x,y)
```

```python
'''if we want the line to be a blue, 
dashed line rather than a red, solid line'''
plt.plot(x,y, color='blue', linestyle = '--')

# The following line of code will generate the same graph as above:

plt.plot(x,y, color = 'b', linestyle = 'dashed')
```

# Sin( ) and cosin( ) graph

Let's create a new graph with the following data points:

X is 200 equally spaced points from - 𝑝𝑖
  to  𝑝𝑖
  (to include  𝑝𝑖
 , we will use the endpoint=True parameter)
the plot will depict two lines, the sine and cosine graphs as functions of X
both lines will be plotted on the same graph

```python
x_sincon = np.linspace(-np.pi, np.pi, 200, endpoint=True)

cosine, sine = np.cos(x_sincon), np.sin(x_sincon)

plt.plot(x_sincon, cosine) # plotting cosine function, red line
plt.plot(x_sincon, sine)   # plotting sine function, blue line
```

```python
# This way will work too:
plt.plot(x_sincon, cosine, x_sincon, sine)
```

```python
# We can make the plot a little bigger
# and we can change the style of the lines
plt.figure(figsize=(10,6))

'''plt.figure(): This function creates a new figure. A figure in Matplotlib can be thought of as a container that holds all elements of the plot such as axes, labels, and other graphical elements.'''

plt.plot(x_sincon, cosine, linestyle='dashed') # plotting cosine function, red dashed line
plt.plot(x_sincon, sine, linestyle='dotted')   # plotting sine function, blue dotted line
```

```python
'''If we don't provide an X axis, matplotlib will use the default X values, starting from 0. 
We can also simply use '--' for a dashed line and ':' for a dotted line:'''
plt.figure(figsize=(10,6))

plt.plot(cosine, '--')
plt.plot(sine, ':')
```

## Adding parameters

We can also add the following to the plot:

Labels for the X and Y axes
Title of the plot
A legend for the lines
You will also see below that as we start adding parameters describing the lines, like labels or colours, we have to use a separate plot() for each of the lines.

```python
plt.figure(figsize=(10,6))

plt.plot(x_sincon, cosine, label = 'cosine')# adding label parameter(shows in the legend if the legend is included)
plt.plot(x_sincon, sine, label = 'sine', color = 'green')    # adding the color parameter and label

plt.ylabel("Y values") # title of the Y axis
plt.xlabel("X values") # title of the X axis
plt.title("Sine and Cosine functions")      # plot title, will be displayed above the plot
plt.legend(loc='upper left')
```

Showing π on a graph

Let's show pi values on the X axis instead of the current integers from -3 to 3. We will also label the X ticks using LaTeX formatting. Finally, we will combine the color and linestyle parameters into a single parameter for simplicity.

```python
plt.figure(figsize=(10, 6))

plt.plot(x_sincon, cosine, '--r', label='cos(x)') # -- for dashed line and r for red
plt.plot(x_sincon, sine, ':g', label='sin(x)', color='green') #qn: Explain :g and color='green'

plt.ylabel("Y values")
plt.xlabel("X values")
plt.title("Sine and Cosine functions")

plt.legend(loc='upper left')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi, np.pi/2], 
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])    #qn: LaTex formatting?

plt.show()
```

# Histograms hist( )

In our previous example, one of the histograms was plotted with sigma = 10 and the other one, with sigma = 30. Let's plot both histograms on the same plot for better visualization.

First, we define the main variables we will need:

```python
mu = 100        # mean of distribution
sigma10 = 10    # standard deviation of distribution, first value of sigma = 10
sigma30 = 30    # standard deviation of distribution, second value of sigma = 30

# Creating two arrays of 10,000 random numbers each with the same mean = 100
# but different standard deviations:
x10 = np.random.normal(loc=mu, scale=sigma10, size=10000)
x30 = np.random.normal(loc=mu, scale=sigma30, size=10000)

'''loc : float or array_like of floats
    Mean ("centre") of the distribution.'''

'''scale : float or array_like of floats
    Standard deviation (spread or "width") of the distribution. Must be non-negative.'''

# Another way of getting the same result as the lines of code above:
# x10 = mu + sigma10 * np.random.randn(10000)
# x30 = mu + sigma30 * np.random.randn(10000)
```

```python
# Size of the histogram plot:
plt.figure(figsize=(20,10))

# Number of bins to be created
num_bins = 200

n, bins, patches = plt.hist(x30, num_bins, density=1, facecolor='b', alpha=0.5, label='$\sigma=30$')
n, bins, patches = plt.hist(x10, num_bins, density=1, facecolor='green', alpha=0.8, label='$\sigma=10$')


'''matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, *, data=None, **kwargs)

Compute and plot a histogram.'''

'''x(n,) array or sequence of (n,) arrays
Input values, this takes either a single array or a sequence of arrays which are not required to be of the same length.'''

'''bins: int or sequence or str, default: rcParams["hist.bins"] (default: 10)
If bins is an integer, it defines the number of equal-width bins in the range.'''

'''the density parameter controls whether the histogram should display the counts or the probability density function (PDF) of the data.

When density is set to False (the default), the histogram bars represent the counts of data points falling into each bin. In this case, the area of all bars sums up to the total number of data points.

When density is set to True, the histogram bars represent the normalized counts, such that the area of the bars sums up to 1. '''

'''The alpha parameter takes a value between 0 and 1, where 0 is completely transparent (invisible) and 1 is completely opaque (not transparent).'''

'''Returns:
narray or list of arrays
The values of the histogram bins.

bins: array
The edges of the bins. 

The patches return value is a list of Patch objects. These Patch objects represent the individual rectangular bars or patches of the histogram.'''

plt.title("Distributions with $\mu=100$, and std deviations $\sigma=10$ and $\sigma=30$")
plt.xlabel("X")
plt.ylabel("Frequencies")
plt.legend()

# Tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.5)

'''matplotlib.pyplot.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)

Adjust the spacing around subplots within a figure. It allows you to customize the layout of the subplots by specifying the amount of space to leave between the edges of the figure and the subplots.

left: float, optional
The position of the left edge of the subplots, as a fraction of the figure width.'''

plt.show()

'''The legend is positioned at the top right of the graph because by default, Matplotlib tries to find the best possible location to place the legend without obstructing the main plot. In this case, it seems that the top right corner was deemed as the most suitable position.'''
```

# Saving graphs savefig( )

In the example below, we are generating a line graph based on the formula  𝑦=−𝑥²−7𝑥+45, where X is an array of 100 data points from 1 to 5, inclusive. We also use the patches library that will allow us to customize the legend of the graph (displayed in the top right corner).

```python
import matplotlib.patches as mpatches

plt.figure(figsize=(10,6))

X = np.linspace(1, 5, 100, endpoint=True)
Y = -X**2 -7*X + 45

plt.plot(X, Y, color='b')

'''matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)[source]
Plot y versus x as lines and/or markers.

Other parameters
**kwargsLine2D properties, optional
kwargs are used to specify properties like a line label (for auto legends), linewidth, antialiasing, marker face color.

color or c
    for color'''

plt.title("Plot for $y = -x^2 - 7x + 45$")
plt.xlabel("X")
plt.ylabel("Y")

# Make sure that we display the grid on the plot
plt.grid(True)

# The legend patch in the top right corner
blue_patch = mpatches.Patch(color="blue", label="$y = -x^2 - 7x + 45$")
plt.legend(handles=[blue_patch])

'''matplotlib.pyplot.legend(*args, **kwargs)[source]
Place a legend on the Axes.'''

'''handles: list of (Artist or tuple of Artist), optional
A list of Artists (lines, patches) to be added to the legend.

Other Parameters:
loc: str or pair of floats, default: rcParams["legend.loc"] (default: 'best')
The location of the legend. 

The string 'best' places the legend at the location, among the nine locations defined so far, with the minimum overlap with other drawn artists. '''

# This will save the graph as PDF in the same folder where the notebook is
plt.savefig("pic1.pdf")
plt.show()
```

# set( )

Here is how we can create the previous graph using set() to define all parameters for axes.

```python
plt.figure(figsize=(10,6))

x_formula = np.linspace(1, 5, 100, endpoint=True)
y_formula = -x_formula ** 2 - 7 * x_formula + 45

ax = plt.axes()

ax.plot(x_formula, y_formula, 'b') #! discouraged ?

# confiugre properties of an axes
ax.set(xlabel='X', ylabel='Y', 
       title='"Plot for $y = -x^2 - 7x + 45$"') 

'''In this context, `set()` is a **method**. Specifically, it is a method associated with the Matplotlib **axes object** (`ax`). When you call `ax.set()`, you can configure multiple properties of the axes simultaneously, such as axis labels, limits, titles, and more. It's a convenient way to customize your plot settings. 📊👍'''

plt.grid(True)

blue_patch = mpatches.Patch(color="blue", label="$y = -x^2 - 7x + 45$")
plt.legend(handles=[blue_patch])
```

Subplots subplots( )

subplots() creates a figure and a set of subplots.

Subplots can be created with the subplot() function, which takes 3 arguments:

Number of rows
Mumber of columns of the subplot grid
Index of the plot to be created
The subplots() function returns figure and axes objects.

Here is a simple example where we will draw the sine and cosine functions on two separate subplots:

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(8,4)) # one row and two columns of subplots

'''matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, width_ratios=None, height_ratios=None, subplot_kw=None, gridspec_kw=None, **fig_kw)

Create a figure and a set of subplots.

**fig_kw
All additional keyword arguments are passed to the pyplot.figure call.'''

'''subplots() returns the following:
fig: Figure

ax: Axes or array of Axes
ax can be either a single Axes object, or an array of Axes objects if more than one subplot was created. '''

'''sharex, sharey: bool or {'none', 'all', 'row', 'col'}, default: False
Controls sharing of properties among x (sharex) or y (sharey) axes:

True or 'all': x- or y-axis will be shared among all subplots.
False or 'none': each subplot x- or y-axis will be independent.'''

ax1.plot(x_sincon, cosine, '--r', label='cos(x)')
ax1.set_title('Sharing Y axis')
ax2.plot(x_sincon, sine, ':g', label = 'sin(x)', color = 'green')

# If we do not provide the number of columns and rows for the subplots() function, matplotlib assumes that we want to use a single row with one column, hence plotting both lines together.
```

Plot on the same graph

If we do not provide the number of columns and rows for the subplots() function, then matplotlib assumes that we want to use a single row with one column, hence plotting both lines together.

```python
f, ax = plt.subplots(figsize=(10,6)) # one row and two columns of subplots

ax.plot(x_sincon, cosine, '--r', label = 'cos(x)')
ax.plot(x_sincon, sine, ':g', label='sin(x)', color = 'green')
ax.legend(frameon=False, loc='upper left')

'''Other Parameters:
loc: str or pair of floats, default: rcParams["legend.loc"] (default: 'best')
The location of the legend.

frameonbool, default: rcParams["legend.frameon"] (default: True)
Whether the legend should be drawn on a patch (frame).'''

# setting plot title and X axes ticks
ax.set(title='Sine and cosine functions',
       xticks= [-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       xticklabels=[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']);
```
